```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#it tells the notebook to display any generated plots or graphs directly in the notebook itself, rather than in a separate window or file. 
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional method
```

```python
plt.plot(x,y)
#plt.show()
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
#object oriented
fig =plt.figure()
axes=fig.add_axes([0.1,0.1,0.5,0.5])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
#The first digit (0.1) represents the left position of the axes within the figure. It's specifying that the left edge of the axes should be 10% from the left edge of the figure.
#The second digit (0.1) represents the bottom position of the axes within the figure. It's specifying that the bottom edge of the axes should be 10% from the bottom edge of the figure.
#The third digit (0.8) represents the width of the axes. It's specifying that the axes should take up 80% of the total width of the figure.
#The fourth digit (0.5) represents the height of the axes. It's specifying that the axes should take up 50% of the total height of the figure.
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,'r')
axes1.set_title('larger plot')
axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('first')
axes[1].plot(y,x)
axes[1].set_title('second')
#axes.plot(x,y)
```

```python
#nrows: This parameter specifies the number of rows of subplots you want to create. It determines how many rows of subplots will be arranged vertically in your figure.

#ncols: This parameter specifies the number of columns of subplots you want to create. It determines how many columns of subplots will be arranged horizontally in your figure.
```

## Figure size DPI (DPI= dots per inch)

```python
fig = plt.figure(figsize=(8,2))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,2))
plt.tight_layout()

axes[0].plot(x,y)
axes[1].plot(y,x)
```
